```python
import pandas as pd
```

```python
country = pd.read_csv('/content/________.csv')
```

```python
df = country

# 취업 가능성 우선순위 지정 (사전 매핑)
priority_map = {"높음": 0, "중간": 1, "낮음": 2}
df["취업_우선순위"] = df["취업 가능성"].map(priority_map)

# 정렬: 취업 가능성 → 소득세율 오름차순 → 최저임금 내림차순
df_sorted = df.sort_values(
    by=["취업_우선순위", "소득세율(%)", "최저임금(USD/월)"],
    ascending=[True, True, False]
).drop(columns=["취업_우선순위"])  # 정렬용 컬럼 제거
```

```python
info = pd.read_csv('/content/국가별_취업정보_줄글형_포함.csv')
```

```python
info
```

```python
summary = pd.merge(country, info, on='국가', how='inner')
summary = summary.drop("취업_우선순위", axis=1)
```

```python
summary
```

```python
df = summary

# 2. 텍스트 → 정형 변수 추출 함수 정의

def extract_visa_features(text):
    return {
        "무비자_가능": "무비자" in str(text),
        "장기비자_필요": "장기" in str(text) or "장기취업" in str(text),
        "비자종류_다양": "비자 종류" in str(text) or "유형" in str(text) or "여러" in str(text),
    }

def extract_industry_features(text):
    keywords = ["제조", "IT", "금융", "물류", "관광", "에너지", "건설", "농업"]
    return {f"수요_{kw}": (kw in str(text)) for kw in keywords}

def extract_social_security_features(text):
    return {
        "건강보험": "건강보험" in str(text),
        "국민연금": "국민연금" in str(text) or "MPF" in str(text),
        "고용보험": "고용보험" in str(text),
        "산재보험": "산재보험" in str(text)
    }

def extract_housing_features(text):
    return {
        "보증금_요구": "보증금" in str(text),
        "월세_선납": "선납" in str(text) or "선불" in str(text),
        "외국인_임대제한": "외국인" in str(text) and "제한" in str(text)
    }

# 3. 각 줄글 컬럼에 적용하여 정형 변수 생성
visa_df = df["비자 정보"].apply(extract_visa_features).apply(pd.Series)
industry_df = df["주요 산업별 수요"].apply(extract_industry_features).apply(pd.Series)
social_df = df["사회보장제도"].apply(extract_social_security_features).apply(pd.Series)
housing_df = df["주택 정보"].apply(extract_housing_features).apply(pd.Series)

# 4. 포함하고 싶은 기존 컬럼들
base_columns = [
    "국가", "대륙", "주요 산업", "비자 유형", "평균 임금 수준",
    "최저임금(USD/월)", "소득세율(%)", "유망 업종", "취업 가능성", "언어 요구사항"
]

# 5. 정형 변수들과 함께 병합
DATA = pd.concat([df[base_columns], visa_df, industry_df, social_df, housing_df], axis=1)
```
