### 매핑

```python
df2 = careerbliss[careerbliss["Company"] == "Apple"]

# 결측치 처리: NaN을 '0/5'로 대체
df2.fillna("0/5", inplace=True)

# 숫자형 변환을 위한 함수 정의
def extract_first_number(value):
    """ '3/5' 같은 형식에서 첫 번째 숫자를 추출하여 변환 """
    return int(float(value.split('/')[0]))

# 변환이 필요한 열들 (정수형으로 변환)
columns_to_convert = [
    "Person You Work For", "People You Work With", "Work Setting", "Support You Get",
    "Rewards You Receive", "Growth Opportunities", "Company Culture", "Way You Work"
]

# 해당 컬럼을 int 형으로 변환
for col in columns_to_convert:
    df2[col] = df2[col].astype(str).apply(extract_first_number).astype(int)

# 'Overall Rating'만 float으로 변환
df2["Overall Rating"] = df2["Overall Rating"].replace("Unknown", "0.0").astype(float)
df2["Overall Rating"] = df2["Overall Rating"].astype(float)

import pandas as pd

# 그룹 키워드 매핑 정의
job_groups = {
"Engineering": [
"engineer", "engineering", "hardware", "software", "developer", "programmer",
"architect", "qa", "sre", "test", "build", "systems", "devops",
"solutions", "hadoop", "machine learning", "technician"
],
"Product / Design": [
"product", "designer", "ui", "ux", "creative", "graphics", "linguistic",
"visual", "image", "media", "design"
],
"Customer Support": [
"advisor", "applecare", "genius", "support", "associate", "agent", "service",
"chat", "help desk", "customer"
],
"Retail / Sales": [
"retail", "store", "cashier", "sales", "representative", "red zone"
],
"Management / PM": [
"manager", "lead", "team manager", "project manager", "program manager",
"coordinator", "supervisor", "director", "relationship"
],
"Marketing": [
"marketing", "brand", "social", "content", "campaign", "Creative Specialist", "specialist"
],
"Finance / Analyst": [
"finance", "accounting", "analyst", "payroll", "controller", "business intelligence",
"quality assurance", "data analyst", "sql"
],
"HR / People": [
"recruiter", "hr", "human resources", "people", "talent", "campus representative"
],
"Internship": [
"intern", "trainee", "apprentice"
],
"IT / Infrastructure": [
"it", "infrastructure", "technical support", "admin", "network", "systems administrator"
],
"Employee" : [
"Apple Employee"
],
"Expert" : [
"Expert", "Technical Expert"
],
"Others": []
}

# 수정된 버전: 대소문자 구분 제거
def categorize_job_title(title):
    if not isinstance(title, str):
        return "Others"

    title_lower = title.lower()

    for category, keywords in job_groups.items():
        if any(keyword in title_lower for keyword in keywords):
            return category

    return "Others"

# 다시 적용
df2["Job Title2"] = df2["Job Title"].apply(categorize_job_title)
```

```python
df2["Job Title2"].value_counts()
```

```python
# 'Job Title2' 열에서 'Others'인 행을 필터링하고, 해당 'Job Title' 값을 추출
job_title_others = df2[df2["Job Title2"] == "Others"]["Job Title"]

# 각 'Job Title' 값들의 빈도를 계산
job_title_counts = job_title_others.value_counts()

job_title_counts
```

### 매핑 수정

```python
# Job Title2가 'Others'인 경우, 'apple employee'와 'technical expert'를 각각 'employee'와 'expert'로 수정
df2.loc[df2['Job Title2'] == 'Others', 'Job Title2'] = df2.loc[df2['Job Title2'] == 'Others', 'Job Title'].apply(
    lambda x: 'Employee' if 'apple employee' in str(x).lower() else ('Expert' if 'technical expert' in str(x).lower() else 'Others')
)
```

```python
df2["Job Title2"].value_counts()
```

```python
grouped_df2 = df2.groupby(["Company", "Job Title2"]).agg({
    "Overall Rating": "mean",  # 평균 계산 (float)
    "Person You Work For": "mean",  # 평균 계산 (int)
    "People You Work With": "mean",
    "Work Setting": "mean",
    "Support You Get": "mean",
    "Rewards You Receive": "mean",
    "Growth Opportunities": "mean",
    "Company Culture": "mean",
    "Way You Work": "mean"
}).reset_index()
```
