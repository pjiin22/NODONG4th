```python
df = careerbliss

# 결측치 처리: NaN을 '0/5'로 대체
df.fillna("0/5", inplace=True)

# 숫자형 변환을 위한 함수 정의
def extract_first_number(value):
    """ '3/5' 같은 형식에서 첫 번째 숫자를 추출하여 변환 """
    return int(float(value.split('/')[0]))

# 변환이 필요한 열들 (정수형으로 변환)
columns_to_convert = [
    "Person You Work For", "People You Work With", "Work Setting", "Support You Get",
    "Rewards You Receive", "Growth Opportunities", "Company Culture", "Way You Work"
]

# 해당 컬럼을 int 형으로 변환
for col in columns_to_convert:
    df[col] = df[col].astype(str).apply(extract_first_number).astype(int)

# 'Overall Rating'만 float으로 변환
df["Overall Rating"] = df["Overall Rating"].replace("Unknown", "0.0").astype(float)
df["Overall Rating"] = df["Overall Rating"].astype(float)

# 그룹화
grouped_df = df.groupby("Company").agg({
    "Job Title": list,  # 리스트로 묶기
    "Overall Rating": "mean",  # 평균 계산 (float)
    "Person You Work For": "mean",  # 평균 계산 (int)
    "People You Work With": "mean",
    "Work Setting": "mean",
    "Support You Get": "mean",
    "Rewards You Receive": "mean",
    "Growth Opportunities": "mean",
    "Company Culture": "mean",
    "Way You Work": "mean"
}).reset_index()
```
